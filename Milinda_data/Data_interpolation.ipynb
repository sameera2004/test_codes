```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib notebook

from scipy.interpolate import interp1d

data = np.loadtxt("FeO-Fe3O4_NP_ga039-00006.chi", skiprows=4)
data.shape
```

```python
plt.figure()
plt.plot(data[:,0], data[:,1], 'o')
plt.show()
```

```python
x_data = np.concatenate((data[1150:1205, 0], data[1215:1265, 0]), axis=0)
y_data = np.concatenate((data[1150:1205, 1], data[1215:1265, 1]), axis=0)

f_d = interp1d(x_data, y_data, bounds_error=False, fill_value=0.,kind='cubic' ) # intreploation
```

```python
xt = np.linspace(244.69,268.94 , 115)
y1 = np.concatenate((data[0:1206, 1], f_d(xt)[56:64]), axis=0)
y = np.concatenate((y1, data[1214::,1]), axis=0)
y.shape
```

```python
plt.figure()
plt.plot(data[:, 0], data[:, 1], '-b', data[:,0], y, '-r')
plt.legend(["data", "i_data"],loc='best')
plt.show()
```
